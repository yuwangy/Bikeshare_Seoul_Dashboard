### Data Loading

```python
import pandas as pd

import chardet

# Detect the encoding
with open('/Users/flashrunner/Bikeshare_Seoul_Dashboard/data/SeoulBikeshareData.csv', 'rb') as file:
    result = chardet.detect(file.read())
    detected_encoding = result['encoding']
    print(f"Detected file encoding: {detected_encoding}")

# Use the actual detected encoding from chardet
df = pd.read_csv('/Users/flashrunner/Bikeshare_Seoul_Dashboard/data/SeoulBikeshareData.csv', encoding=detected_encoding)

# Check the first few rows of the DataFrame
df.head(10)
```

### Data Wrangling & Cleaning

```python
# Create the Weekend, Month column

# Add 'Month' as ordinal attribute
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')

# Create 'Month' column based on date
df['Month'] = df['Date'].dt.month

# Add 'Weekend' as nominal attribute, determine if the date is on a weekend 
df['Weekend'] = df['Date'].dt.dayofweek.apply(lambda x: 'Yes' if x >= 5 else 'No')

# Check the first few rows
df.head()
```
